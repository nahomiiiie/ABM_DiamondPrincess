## Formulas for Editing

```python
def minimizeindex(df):
    df = df[df['x'] % 200 == 0]
    return (df)
```

```python
def avg(df):
    m = df.y.mean()
    return(m)
```

## Incorporating dataframes

```python
allcont = {}
daily = []

allcont[0] = df0
allcont[1] = df1
allcont[2] = df2
allcont[3] = df3
allcont[4] = df4
allcont[5] = df5
allcont[6] = df6
allcont[7] = df7
allcont[8] = df8
allcont[9] = df9

allcont[10] = df10
allcont[11] = df11
allcont[12] = df12
allcont[13] = df13
allcont[14] = df14
allcont[15] = df15
allcont[16] = df16
allcont[17] = df17
allcont[18] = df18
allcont[19] = df19
allcont[20] = df20
```

```python

for item in allcont:
    thisdf = allcont[item]
    allcont[item] = minimizeindex(thisdf)
    daily.append(avg(thisdf))
    

```

```python
np.mean(daily)
```

```python

n, bins, patches = plt.hist(daily, 10, color='c', edgecolor = 'black', alpha=0.7)
plt.axvline(np.mean(daily), color='red', linestyle='dashed', linewidth=1)
min_ylim, max_ylim = plt.ylim()
plt.text(np.mean(daily) + .01, 3.5, 'Mean: 39.57')
plt.title('Average Daily Contacts Across Model Runs')
plt.xlabel('Number of Contacts')
plt.ylabel('Runs')
plt.yticks([0, 1, 2, 3, 4])


plt.savefig('../../plots/AvgContacts.png', transparent = True)
plt.show()
```

```python
for group in allcont:
    plt.plot(allcont[group].x[1:], allcont[group].y[1:])

plt.axvline(2000, color='black', linestyle='solid', linewidth=3)
plt.text(2050, 42.7, 'Passenger/Infected Quarantine at Day 10')
plt.title('Contacts per Day')
plt.xlabel('Ticks')
plt.ylabel('Number of Contacts')
plt.xticks(range(0,6200,500))
plt.savefig('../../plots/ContactPerDay500.png', transparent=True)
plt.show()
```
